#### Train mnist_cnn1:

```python
from __future__ import print_function

import os #avoid importing the entire module
import keras #avoid importing the entire module
from keras.models import Sequential, load_model
from keras.layers import Activation, Dense, Dropout, Flatten, Conv2D, MaxPooling2D

## checkpointer callback imports ##
###################################
from keras.callbacks import ModelCheckpoint, TensorBoard
from pipeline.keras.callbacks import LossHistory

model_id = "visualize"

x_train = train[0]
y_train = train[1]
x_test = test[0]
y_test = test[1]
batch_size = 128
num_classes = 10
epochs = 1

img_rows, img_cols = 28, 28
input_shape = (img_rows, img_cols, 1)

model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),
                 input_shape=input_shape))
convout1 = Activation('relu')
model.add(convout1)
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128))
cnn_codes = Activation('relu')
model.add(cnn_codes)
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))
model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])

# If model directory does not exist, make one
model_name = "mnist_cnn_{}".format(model_id)
model_dir = "./results/" + model_name
checkfile = "/chckpt_weights.{epoch:02d}-{val_loss:.2f}.hdf5"
filepath = model_dir + checkfile
if not os.path.exists(model_dir):
                os.makedirs(model_dir)

## Callbacks
loss_history = LossHistory()

# Warning: don't use .format() method {} conflicts with keras parsing, tuple out of bounds error
checkpointer = ModelCheckpoint(filepath=filepath, verbose=1, save_best_only=True)
history = model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs,
          verbose=0, validation_data=(x_test, y_test), callbacks = [checkpointer, loss_history])

# saves model as an HDF5 file
model.save(model_dir+"/{}".format(model_name) + ".h5")

score = model.evaluate(x_test, y_test, verbose=0)
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
import numpy as np
import matplotlib.pyplot as plt

model.summary()
losses = loss_history.losses
plt.plot(np.linspace(0,len(losses)/2, len(losses)/2), losses[1::2], color='lightblue')
plt.xlabel("Batch Number")
plt.ylabel("Loss")
plt.title("Figure 1: Loss Function on clean data")
plt.show()
```

```python
import numpy as np
from sklearn.metrics import classification_report

# Test set error rate
loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
pred_classes = model.predict_classes(x_test, verbose=0)
true_labels = np.argmax(y_test, axis=1)

# Per class error rate per class (This has a bug)
nclasses = 10
class_accuracies = np.zeros(nclasses)
class_totals = np.zeros(nclasses)
class_correct = np.zeros(nclasses)
for class_id in range(nclasses):
    for i in range(len(true_labels)):
        if class_id == true_labels[i]:
            class_totals[class_id] += 1
            if true_labels[i] == pred_classes[i]:
                class_correct[class_id] += 1
class_accuracies = np.divide(class_correct, class_totals)

error_rate = 1 - accuracy
error_rates = 1 - class_accuracies

print("The error rate for each class is: ", np.round(error_rates, 3))
print("The test error rate is: {}".format(np.round(error_rate, 3)))
print(classification_report(true_labels, pred_classes))
```

#### Designing an Experiment to Confidently Bound Label Corruption in Large Datasets

For a 98% $(1-\alpha)\%$ we have $\frac{\alpha}{2} = 0.01$. This gives us a z score of 2.33. If our population $N$ the size of the training set and we desire an error $\epsilon < 0.01$ on our estimater $\hat{p}$, then we will need a sample size that is at least $n=\dfrac{m}{1+\dfrac{m-1}{N}}$, where $m=\dfrac{z^2_{\alpha/2}\hat{p}(1-\hat{p})}{\epsilon^2}$ is the upper bound on the sample size in the limit that $N >> m$.

```python
# Calculate the minimum sample size for the specified uncertainties
# in the sample mean given a specified confidence conf

import scipy.stats as st
import numpy as np

# load MNIST
(input_shape, (x_train, y_train),(x_test, y_test)) = formatted_mnist()

# Price in dollars to label an image. 
# Sources: http://labelme.csail.mit.edu/Release3.0/browserTools/php/mechanical_turk.php - $0.1/5 per label quoted
# http://cloudacademy.com/blog/machine-learning-datasets-mechanical-turk/ $0.20 - $0.30 per label quoted
# https://aws.amazon.com/blogs/aws/new-mechanical-turk-resource-center/
price_per_label = 0.2
conf = 0.99 # Specify desired confidence
print("Confidence: {}%".format(conf*100) + "\n")

corruption_uncertainty = [0.25, 0.20, 0.15, 0.10, 0.05, 0.01]
for epsilon in corruption_uncertainty:
    z_score = st.norm.ppf(conf) # compute alpha/2 z-score
    population_size = x_train.shape[0]
    sample_mean_param = 0.5 #Conservative choice: maximum entropy
    sample_limit = (z_score)**2*sample_mean_param*(1-sample_mean_param)/(epsilon**2)
    min_sample_size = sample_limit/(1+(sample_limit-1)/population_size)
    #print("sample_mean: ", sample_mean)
    print("if sample_limit applies: ", int(np.ceil(sample_limit)))
    print("min_sample_size: ", int(np.ceil(min_sample_size)))
    print("Specified error: ", epsilon)
    print("Cost of precision: ${}".format(np.round(price_per_label*min_sample_size, 2)) + "\n")
```
